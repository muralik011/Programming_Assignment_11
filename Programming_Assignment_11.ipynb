1. Write a Python program to find words which are greater than given length k?

```python
l = input('Enter words separated by space: ')
l = l.split()
k = int(input('Enter the threshold length: '))
new_l = [i for i in l if len(i) > k]
print(new_l)
```

2. Write a Python program for removing i-th character from a string?

```python
s = input('Enter a string: ')
s = list(s)
i = int(input('Enter \'i\' as in ith character to remove from string: '))
s = ''.join([j for n, j in enumerate(s) if n != (i - 1)])
print(s)
```

3. Write a Python program to split and join a string?

```python
s = input('Enter a string: ')
s_split = list(s)
s_joined = ''.join(s_split)
print(s_split)
print(s_joined)
```

4. Write a Python to check if a given string is binary string or not?

```python
import re
s = input('Enter a string: ')
s = re.sub(pattern='[01]', repl='', string=s)
if len(s) == 0:
    print('string is binary')
else:
    print('string is not binary')
```

5. Write a Python program to find uncommon words from two Strings?

```python
s = input('Enter a sentence: ')
s1 = input('Enter another sentence: ')
s = set(s.lower().split())
s1 = set(s1.lower().split())
print(s.difference(s1))
```

6. Write a Python to find all duplicate characters in string?

```python
s = input('Enter a string: ')
duplicate_chars = [*set([i for i in s if s.count(i) > 1])]
print(duplicate_chars)
```

7. Write a Python Program to check if a string contains any special character?

```python
s = input('Enter a string: ')
s_without_sp_chars = re.sub(pattern='[^\w\s]', repl='', string=s)
if len(s_without_sp_chars) < len(s):
    print(f'string contains special characters {set(s).difference(set(s_without_sp_chars))}')
else:
    print('string has no special characters')
```
